```python
#Importar as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#Importar os dados
dados = pd.read_csv('dados.csv', sep=';')
dados.head()
```

```python
dados.shape
```

```python
#Procurar por dados nulos
dados.isnull().sum()
```

```python
#Visualizar o PIB
srn.distplot(dados.PIB, bins = 10, axlabel='PIB').set_title('PIB')
```

```python
agrupado = dados.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='PIB')
```

```python
#Visualizar o DESEMPENHO
srn.distplot(dados.VALOREMPENHO, bins = 10, axlabel='Empenho').set_title('Empenho')
```

```python
agrupado = dados.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO')
```
